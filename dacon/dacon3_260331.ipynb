```python
# 필수 import
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
# 데이터 불러오기
train = pd.read_csv('/content/train.csv', index_col=0)
test = pd.read_csv('/content/test.csv', index_col=0)
sample_submission = pd.read_csv('/content/sample_submission.csv', index_col=0)
```

train 데이터의 type을 sample_submission에 대응하는 가변수 형태로 변환시킵니다.

```python
column_number = {}
for i, column in enumerate(sample_submission.columns):
    column_number[column] = i

def to_number(x, dic):
    return dic[x]

train['type_num'] = train['type'].apply(lambda x: to_number(x, column_number))
```

모델에 적용하기 위한 데이터셋


```python
train_x = train.drop(columns=['type', 'type_num'], axis=1)
train_y = train['type_num']
test_x = test
```

모델 훈련

```python

forest = RandomForestClassifier(n_estimators=100, n_jobs=-1)
forest.fit(train_x, train_y)
```

생성된 모델로 예측 데이터 생성

```python
y_pred = forest.predict_proba(test_x)
```

```python
# submission


submission = pd.DataFrame(data=y_pred, columns=sample_submission.columns, index=sample_submission.index)
submission.to_csv('submission.csv', index=True)
```

LightGBM + StratifiedKFold

```python
import numpy as np
import pandas as pd
import lightgbm as lgb

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import log_loss
```

```python
n_classes = y.nunique()

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

oof_pred = np.zeros((len(X), n_classes))
test_pred = np.zeros((len(X_test), n_classes))

for fold, (train_idx, valid_idx) in enumerate(skf.split(X, y), 1):
    X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
    y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

    model = lgb.LGBMClassifier(
        objective='multiclass',
        num_class=n_classes,
        n_estimators=500,
        learning_rate=0.03,
        num_leaves=31,
        max_depth=-1,
        min_child_samples=20,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    )

    model.fit(
        X_train, y_train,
        eval_set=[(X_valid, y_valid)],
        eval_metric='multi_logloss',
        callbacks=[
            lgb.early_stopping(50),
            lgb.log_evaluation(100)
        ]
    )

    oof_pred[valid_idx] = model.predict_proba(X_valid)
    test_pred += model.predict_proba(X_test) / skf.n_splits

    fold_score = log_loss(y_valid, oof_pred[valid_idx])
    print(f'Fold {fold} LogLoss: {fold_score:.6f}')
```

OOF 점수 확인

```python
oof_score = log_loss(y, oof_pred)
print(f'OOF LogLoss: {oof_score:.6f}')
```

```python
submission = sample_submission.copy()
# test_pred의 19개 열을 sample_submission의 19개 클래스 열에 직접 할당합니다.
submission[sample_submission.columns] = test_pred
submission.to_csv('submission_oof_lgbm.csv', index=False)

print(submission.head())
print(submission.iloc[:, :].sum(axis=1).head())
```
