```python
# train과 test 데이터프레임을 합쳐서 전처리 과정을 일관되게 적용합니다.
df = pd.concat([train.drop('box_off_num', axis=1), test], ignore_index=True)

# release_time 컬럼을 datetime 객체로 변환합니다.
df['release_time'] = pd.to_datetime(df['release_time'])

# 연, 월, 일, 요일, 시간 피처를 추출합니다.
df['release_year'] = df['release_time'].dt.year
df['release_month'] = df['release_time'].dt.month
df['release_day'] = df['release_time'].dt.day
df['release_dayofweek'] = df['release_time'].dt.dayofweek
df['release_hour'] = df['release_time'].dt.hour

# 원본 release_time 컬럼은 더 이상 필요 없으므로 제거합니다.
df = df.drop('release_time', axis=1)

# dir_prev_bfnum 컬럼의 결측값을 0으로 채웁니다.
df['dir_prev_bfnum'] = df['dir_prev_bfnum'].fillna(0)

display(df.head())
```

위 코드에서는 `train`과 `test` 데이터프레임을 합쳐 `release_time` 컬럼에서 연, 월, 일, 요일, 시간 피처를 추출했습니다. 또한 `dir_prev_bfnum` 컬럼의 결측값은 0으로 채웠습니다. 이제 범주형 변수를 처리할 차례입니다.

```python
from sklearn.preprocessing import LabelEncoder

# 범주형 컬럼을 식별합니다.
categorical_features = ['distributor', 'genre', 'screening_rat', 'director']

# 각 범주형 컬럼에 대해 Label Encoding을 수행합니다.
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

display(df.head())
```

위 코드에서는 `distributor`, `genre`, `screening_rat`, `director`와 같은 범주형 컬럼에 `LabelEncoder`를 적용하여 수치형 데이터로 변환했습니다. 이제 데이터를 다시 `train`과 `test` 세트로 분리하고 모델 학습을 위한 준비를 할 수 있습니다.

```python
# 전처리된 데이터를 다시 train과 test로 분리합니다.
X_train = df.iloc[:len(train)]
X_test = df.iloc[len(train):]
y_train = train['box_off_num']

display(X_train.head())
display(X_test.head())
```

```python
import lightgbm as lgb

# LightGBM 모델 정의
model = lgb.LGBMRegressor(random_state=42)

# 모델 학습
model.fit(X_train, y_train)

# 예측 수행
predictions = model.predict(X_test)

display(predictions[:5])
```

위 코드에서는 `LGBMRegressor` 모델을 초기화하고, `X_train`과 `y_train`을 사용하여 모델을 학습시켰습니다. 학습된 모델은 `X_test` 데이터에 대한 예측을 생성했으며, 상위 5개의 예측값을 출력했습니다. 이제 예측 결과를 submission 파일 형식에 맞게 저장하겠습니다.

```python
# submission 파일에 예측 결과 저장
submission['box_off_num'] = predictions

# 예측값이 음수가 되는 경우 0으로 처리 (관객수는 음수가 될 수 없으므로)
submission['box_off_num'] = submission['box_off_num'].apply(lambda x: 0 if x < 0 else x)

# 결과 파일 저장
submission.to_csv('submission_lgbm.csv', index=False)

display(submission.head())
```

LightGBM 모델을 사용하여 예측을 완료하고, 예측 결과를 `submission_lgbm.csv` 파일로 저장했습니다. 예측값이 음수인 경우 0으로 처리하는 로직도 추가했습니다. 다음 단계는 K-Fold 교차 검증을 적용하거나 하이퍼파라미터 튜닝을 진행하는 것입니다.

```python
from sklearn.model_selection import KFold
import numpy as np

# K-fold 설정
NFOLDS = 5
फोल्डर = KFold(n_splits=NFOLDS, shuffle=True, random_state=42)

# K-fold 예측을 저장할 배열 초기화
kfold_predictions = np.zeros(X_test.shape[0])

# K-fold 모델 학습 및 예측
for fold_, (trn_idx, val_idx) in enumerate(फोल्डर.split(X_train, y_train)):
    print(f"Current Fold: {fold_}")
    trn_data = X_train.iloc[trn_idx]
    trn_target = y_train.iloc[trn_idx]

    val_data = X_train.iloc[val_idx]
    val_target = y_train.iloc[val_idx]

    # LightGBM 모델 정의 (기본 파라미터 사용)
    model = lgb.LGBMRegressor(random_state=42)

    # 모델 학습
    model.fit(trn_data, trn_target)

    # 예측 수행
    kfold_predictions += model.predict(X_test) / NFOLDS

display(kfold_predictions[:5])
```

위 코드에서는 5-겹 K-Fold 교차 검증을 사용하여 LightGBM 모델을 학습하고 예측을 수행했습니다. 각 폴드마다 모델을 훈련하고 `X_test`에 대한 예측을 생성한 후, 이 예측값들을 평균하여 최종 K-Fold 예측값을 도출했습니다. 이는 모델의 안정성과 일반화 성능을 높이는 데 도움이 됩니다.

이제 이 K-Fold 예측 결과를 submission 파일에 저장하겠습니다.

```python
# submission 파일에 K-Fold 예측 결과 저장
submission_kfold = submission.copy() # 원본 submission 데이터프레임을 복사하여 사용
submission_kfold['box_off_num'] = kfold_predictions

# 예측값이 음수가 되는 경우 0으로 처리
submission_kfold['box_off_num'] = submission_kfold['box_off_num'].apply(lambda x: 0 if x < 0 else x)

# 결과 파일 저장
submission_kfold.to_csv('submission_lgbm_kfold.csv', index=False)

display(submission_kfold.head())
```

K-Fold 교차 검증을 통해 얻은 예측값을 `submission_lgbm_kfold.csv` 파일로 저장했습니다. 다음 단계는 Grid Search를 통한 하이퍼파라미터 튜닝을 진행하거나, 추가적인 Feature Engineering을 고려해 볼 수 있습니다.

```python
from sklearn.model_selection import GridSearchCV

# LightGBM 모델 정의
lgbm = lgb.LGBMRegressor(random_state=42)

# 튜닝할 하이퍼파라미터 범위 설정
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'num_leaves': [20, 31, 40],
    'max_depth': [-1, 10, 20]
}

# GridSearchCV 객체 생성
grid_search = GridSearchCV(estimator=lgbm, param_grid=param_grid, cv=KFold(n_splits=5, shuffle=True, random_state=42), scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)

# 그리드 서치 수행
grid_search.fit(X_train, y_train)

# 최적의 하이퍼파라미터 및 최고 점수 출력
print("Best parameters found: ", grid_search.best_params_)
print("Lowest RMSE found: ", np.sqrt(-grid_search.best_score_))
```

위 코드에서는 `GridSearchCV`를 사용하여 LightGBM 모델의 최적 하이퍼파라미터를 탐색했습니다. `n_estimators`, `learning_rate`, `num_leaves`, `max_depth`에 대한 다양한 조합을 5-Fold 교차 검증을 통해 평가하고, 가장 낮은 RMSE를 달성하는 파라미터 조합을 찾았습니다.

이제 이 최적의 파라미터로 모델을 재학습하고 예측을 수행하겠습니다.

```python
# 최적의 하이퍼파라미터로 모델 재학습
best_lgbm_model = lgb.LGBMRegressor(**grid_search.best_params_, random_state=42)
best_lgbm_model.fit(X_train, y_train)

# 최적화된 모델로 예측 수행
optimized_predictions = best_lgbm_model.predict(X_test)

# submission 파일에 최적화된 예측 결과 저장
submission_optimized = submission.copy()
submission_optimized['box_off_num'] = optimized_predictions

# 예측값이 음수가 되는 경우 0으로 처리
submission_optimized['box_off_num'] = submission_optimized['box_off_num'].apply(lambda x: 0 if x < 0 else x)

# 결과 파일 저장
submission_optimized.to_csv('submission_lgbm_optimized.csv', index=False)

display(submission_optimized.head())
```

Grid Search를 통해 최적화된 LightGBM 모델로 예측을 수행하고, 그 결과를 `submission_lgbm_optimized.csv` 파일로 저장했습니다. 이로써 기본 LightGBM, K-Fold LightGBM, 그리고 하이퍼파라미터 튜닝을 거친 LightGBM 모델 학습 및 예측 과정이 모두 완료되었습니다.
